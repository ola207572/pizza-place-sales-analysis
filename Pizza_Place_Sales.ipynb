Question (6), what are the peak hours of sales

```python
df = order.merge(order_details, on ="order_id")
df
```

```python
df["revenue"]= df["quantity"] * df["price"]
```

```python
df["hour"] = pd.to_datetime(df["time"]).dt.hour
df
```

```python
hourly_sales= df.groupby("hour")["revenue"].sum()
print(hourly_sales)
```

```python
peak_hour = hourly_sales.idxmax()
peak_sales = hourly_sales.max()

print(f"Peak Hour: {peak_hour}:00")
print(f"Sales: ${peak_sales}")
```

Qeustion (7), Find the total sales made on each day of the week. Which day of the week is when sales
are made the most

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["day_of_week"]= df["date"].dt.day_name()
```

```python
sales_by_day= df.groupby("day_of_week")["revenue"].sum()
sales_by_day
```

```python
days = [
    "Monday", "Tuesday", "Wednesday", 
    "Thursday","Friday", "Saturday","Sunday"
]

sales_by_day =sales_by_day.reindex(days)
print(sales_by_day)
```

```python
best_day=sales_by_day.idxmax()
best_sales= sales_by_day.max()

print(best_day)
print(best_sales)
```
